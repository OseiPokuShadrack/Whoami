```python
print ("           STUDENT  PROFILE SYSTEM            "   )
print ("===============================================")
name = input ("Enter Your Full Name:     ")
age_input =  input ("Enter Your Age:      ")
age = int(age_input)
gpa_input = input ("Enter Your GPA:   ")
gpa = float(gpa_input)
(gpa_input)
courses = input ("Enter Your Courses(seprate with comma): ").split(",")

student_profile = {
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}

dean_list = gpa > 3.49
minor_age = age < 18

print (" \n ===== TYPE CONVERSION =====")
print ("Age original type: ", type(age_input)) 
print ("Age converted type: ", type(age))
print ("GPA original type: ", type(gpa_input))
print ("GPA converted type: ", type(gpa))

print (" \n ===== STUDENT PROFILE =====")
print (f"Name: {student_profile['name']}") 

if minor_age:
 print (f"Age: {student_profile['age']} (Minor)") 
else:
 print (f"Age: {student_profile['age']} (Adult)") 

print (f"GPA: {student_profile['gpa']}")
print (f"Courses: {student_profile['courses']}")
print (f"# Courses: {student_profile['num_courses']}")

if dean_list :
 print ("Dean's List : Yes")
else:
 print ("Dean's List : No")

print ("***********************************************************************************************")
```
